```python
#Dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sportsreference.nba.teams import Teams
```

```python
#Importing stats from the 2019-2020 NBA Season *A full regular season was not played due to Covid-19*
# from sportsreference.nba.teams import Teams
teams = Teams("2020")

stats_19_20 = teams.dataframes

stats_19_20.head()
```

```python
#Importing stats from the 2018-2019 Season
teams = Teams("2019")

stats_18_19 = teams.dataframes

stats_18_19.head()
```

```python
#Importing stats from the 2017-2018 season
stats_17_18 = Teams("2018").dataframes

stats_17_18.head()
```

```python
#Importing team wins from the 2017-2018 Season
wins_17_18 = pd.read_csv('../Data/team_wins_17_18.csv')  
wins_17_18.head()
```

```python
#Importing team wins form the 2018-2019 season
wins_18_19 = pd.read_csv('../Data/team_wins_18_19.csv')  
wins_18_19.head()
```

```python
#Importing team wins from the 2019-2020 season
wins_19_20 = pd.read_csv('../Data/team_wins_19_20.csv')  
wins_19_20.head()
```

```python
#Merging team stats with team wins for 2019-2020
stats_19_20 = stats_19_20.merge(wins_19_20, left_on='abbreviation', right_on='abbreviation')
stats_19_20 = stats_19_20.assign(Season=['2019-2020'] * 30)
stats_19_20.head()
```

```python
#Merging team stats for the 2018-2019 season
stats_18_19 = stats_18_19.merge(wins_18_19, left_on='abbreviation', right_on='abbreviation')
stats_18_19 = stats_18_19.assign(Season=['2018-2019'] * 30)
stats_18_19.head()
```

```python
#Mergin team stats for the 2017-2018 season
stats_17_18 = stats_17_18.merge(wins_17_18, left_on='abbreviation', right_on='abbreviation')
stats_17_18 = stats_17_18.assign(Season=['2017-2018'] * 30)
stats_17_18.head()
```

```python
#Concatinating the stats into one dataframe
all_df = [stats_17_18, stats_18_19, stats_19_20]

all_df = pd.concat(all_df).reset_index(drop=True)

all_df.head()
```

```python
#all_df = read_csv("total_stats.csv")
```

```python
# Assign the data to X and y

X = all_df[["abbreviation","Season","assists", 'defensive_rebounds','field_goal_percentage','field_goals','opp_assists',
                        'opp_blocks','opp_defensive_rebounds', 'opp_field_goal_percentage','opp_points',
                        'opp_three_point_field_goal_percentage','opp_total_rebounds','opp_two_point_field_goal_percentage',
                        'points', 'three_point_field_goal_percentage', 'total_rebounds','two_point_field_goal_percentage']].set_index(['abbreviation', 'Season'])
y = all_df["W"].values.reshape(-1, 1)
#Checking the shapes
print(X.shape, y.shape)
```

```python
# Use train_test_split to create training and testing data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Create a LinearRegression model and fit it to the scaled training data

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
predictions = predictions.flatten()

#Printing the model MSE and Score for evaluation
MSE = mean_squared_error(y_test, predictions)
r2 = model.score(X_test, y_test)

print(f"MSE: {MSE}, R2: {r2}")


lin_reg = pd.DataFrame({"Prediction": np.round(predictions , 2), "Actual": y_test.ravel()})

X_test.reset_index(inplace = True)

df_out = pd.merge(X_test, lin_reg, how = 'left',left_index = True, right_index = True)

linear_Model_df = df_out.loc[:,['Season', 'abbreviation', 'Prediction', 'Actual']]
linear_Model_df

#Writing the results to a csv and viewing the output 
linear_Model_df.to_csv('../Data/Linear_Model_Predictions.csv', index = False)

linear_Model_df
```
